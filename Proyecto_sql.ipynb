```python
#GUARDAR LA NUEVA BASE SIN NAN
df = df.dropna () 
```

```python
#VERIFICAR SI HAY NAN NUEVAMENTE
df.isnull (). sum()
```

```python
#CAMBIAR EL TIPO DE DATO DE LAS COLUMNAS, EN ESTE CASO A INTEGER
#df['Release Year'] = df['Release Year'].astype(int)
```

```python
df['Release Year'] = pd.to_numeric(df['Release Year'], errors='coerce').fillna(0).astype(int)
```

```python
df.info ()
```

```python
df['Votes'] = df['Votes'].str.replace(',', '').astype(int)
```

```python
df.info ()
```

```python
df
```

```python
def convert_gross(value):
    # Eliminar el símbolo '$'
    value = value.replace('$', '')
    # Verificar si el sufijo es 'M' (millones) y convertir a número
    if 'M' in value:
        value = value.replace('M', '')
        value = float(value) * 1_000_000
    # Verificar si el sufijo es 'B' (mil millones) y convertir a número
    elif 'B' in value:
        value = value.replace('B', '')
        value = float(value) * 1_000_000_000
    # Convertir el valor a float y luego a int
    return int(float(value))

# Aplicar la función a la columna 'Gross'
df['Gross'] = df['Gross'].apply(convert_gross)
```

```python
df_movie = df[["Movie Name", "Release Year","Duration",	"IMDB Rating",	"Metascore", "Votes", "Genre", "Gross"]].reset_index ()
df_movie
```

```python
df_movie.index.name = 'Movie ID'

# Ajustar el índice para que comience desde 1
df_movie.index = df.index + 1

df_movie
```

```python
df_movie.rename(columns={'index': 'Movie ID'}, inplace=True)
df_movie['Movie ID'] = df_movie['Movie ID'] + 1
df_movie
```
